```python
!pip install ipywidgets

import ipywidgets as widgets
from IPython.display import display

# Define a text widget for input and output
Output = widgets.Text(value="", layout=widgets.Layout(width="200px"))

# Create numeric buttons (0-9) and operator buttons (+, -, *, /, =, C)
buttons = [widgets.Button(description=str(i), layout=widgets.Layout(width="50px")) for i in range(10)]
operators = ["+", "-", "*", "/", "=", "C"]
operator_buttons = [widgets.Button(description=op, layout=widgets.Layout(width="50px")) for op in operators]

# Create button grid layout
button_grid = [buttons[i:i+3] for i in range(0, len(buttons), 3)]
button_grid.append(operator_buttons)

# Define a button click event handler
def on_button_click(b):
    current_text = Output.value
    clicked_text = b.description
    if clicked_text == "=":
        try:
            Result = str(eval(current_text))
            Output.value = Result
        except:
            Output.value = "Error"
    elif clicked_text == "C":
        Output.value = ""
    else:
        Output.value = current_text + clicked_text

# Attach the event handler to the buttons
for row in button_grid:
    for button in row:
        button.on_click(on_button_click)

# Create button layout and calculator layout
button_layout = widgets.VBox([widgets.HBox(row) for row in button_grid])
calculator_layout = widgets.VBox([Output, button_layout])

# Display the calculator interface
display(calculator_layout)
```

```python
#split an email into username and domain
def email_slicer(email):
    try:
        # Split the email address into username and domain parts
        username, domain = email.split('@')
        # Display the result
        print(f"Username: {username}")
        print(f"Domain: {domain}")
    except ValueError:
        print("Invalid email format. Please enter a valid email address.")

if __name__ == "__main__":
    # Get email address input from the user
    email_address = input("Enter your email address: ")
    # Call the email_slicer function with the provided email address
    email_slicer(email_address)
```

```python
#fib series using while loop
n = int(input("Enter the number of terms: "))
a = 0
b = 1
sum = a + b 
count = 1
print("Fibonacci series is:", end="") 
while count <= n:
   count += 1 
   print(a, end=" ")
   a = b
   b = sum
   sum = a + b
```

```python
def fibonacci(n):
    # Initialize the first two terms
    f = [0, 1]
    for i in range(2, n + 1):
        f.append(f[i - 1] + f[i - 2])
    return f

n = int(input("Enter the number of terms: "))
ans = fibonacci(n)

for i in range(n):
    print(ans[i], end=' ')

```
